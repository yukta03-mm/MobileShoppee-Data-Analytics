```python
# Import Required Libraries
import pandas as pd
import numpy as np

# Load All CSV Files
customers = pd.read_csv("Customers.csv")
models = pd.read_csv("MobileModels.csv")
shops = pd.read_csv("ShopLocations.csv")
sales = pd.read_csv("Sales.csv")
stocks = pd.read_csv("Stocks.csv")
offers = pd.read_csv("Offers.csv")

# Preview
print("Data Loaded Successfully!\n")
print("Sales Columns:", sales.columns.tolist())
print("Customers Columns:", customers.columns.tolist())
```

```python
# derive KPIs

sales["revenue"] = sales["quantity"] * sales["final_price"]
total_revenue = sales["revenue"].sum()
print(f"Total Revenue: ₹{total_revenue:,.2f}")
```

```python
avg_discount = sales["discount"].mean()
print(f"🏷 Average Discount: {avg_discount:.2f}%")
```

```python
repeat_customers = sales["customer_id"].value_counts()
num_repeat_customers = (repeat_customers > 1).sum()
total_customers = sales["customer_id"].nunique()
repeat_ratio = (num_repeat_customers / total_customers) * 100
print(f" Repeat Customers: {num_repeat_customers} ({repeat_ratio:.2f}% of all customers)")
```

```python
# Merge sales and stock data
stock_merge = pd.merge(stocks, sales.groupby("model_id")["quantity"].sum().reset_index(), 
                       on="model_id", how="left")

stock_merge["stock_turnover"] = stock_merge["quantity"] / stock_merge["available_units"]
avg_turnover = stock_merge["stock_turnover"].mean()
print(f"Average Stock Turnover: {avg_turnover:.2f}")
```

```python
# Merge offers with sales on common columns: model_id and shop_id
sales_offers = pd.merge(sales, offers, on=["model_id", "shop_id"], how="left")

# Ensure 'revenue' exists
sales_offers["revenue"] = sales_offers["quantity"] * sales_offers["final_price"]

# Group by campaign or offer_id (if present)
roi_df = sales_offers.groupby("campaign_name").agg({
    "revenue": "sum",
    "discount_percent": "mean"
}).reset_index()

# Compute ROI estimate safely
roi_df["roi_estimate"] = roi_df["revenue"] / (roi_df["discount_percent"] + 1e-6)

# Display
print("Campaign ROI (Top 5 offers):")
print(roi_df.head())
```

```python
def detect_outliers(df, column):
    if column in df.columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        outliers = df[(df[column] < lower) | (df[column] > upper)]
        print(f"{column}: {len(outliers)} outliers detected.")
        return outliers
    else:
        print(f"{column} not found in dataset.")
        return pd.DataFrame()

# Check in Sales and Customers
outliers_final_price = detect_outliers(sales, "final_price")
outliers_discount = detect_outliers(sales, "discount")
outliers_loyalty = detect_outliers(customers, "loyalty_points")
```

```python
# Convert column names to snake_case.

#Convert column names in all CSV files within a folder to snake_case format.

import pandas as pd       # For reading and writing CSV files
import re                 # For regular expressions used in renaming columns
import os                 # For interacting with the file system

# Define the folder path containing your CSV files
folder_path = r'C:\Users\katka\OneDrive\ドキュメント\EDA_Project'

# Function: Convert a string to snake_case
def to_snake_case(col):
    """
    Converts a given column name to snake_case format.   
    Parameters:
    col (str): The original column name.
    Returns:
    str: The column name converted to snake_case.
    """
    # Replace any non-alphanumeric characters with underscores
    col = re.sub(r'[^0-9a-zA-Z]+', '_', col)
    
    # Insert underscore between lowercase/number and uppercase letters (e.g., "userID" → "user_ID")
    col = re.sub(r'([a-z0-9])([A-Z])', r'\1_\2', col)
    
    # Convert to lowercase and strip leading/trailing underscores
    return col.lower().strip('_')

# Loop through all files in the specified folder
for filename in os.listdir(folder_path):
    # Process only CSV files
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)  # Full path to the file
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Apply snake_case conversion to all column names
        df.columns = [to_snake_case(col) for col in df.columns]
        
        # Save the updated DataFrame back to the same file (overwrite)
        df.to_csv(file_path, index=False)
        
        # Print confirmation for each updated file
        print(f"Updated: {filename}")
```

```python
# Normalize only date/datetime columns to ISO-8601 (YYYY-MM-DD) without touching numeric columns.
# - No parse_dates on read
# - Detect date-like columns using robust heuristics (avoid numeric)
# - Convert selected columns using pandas.to_datetime(errors='coerce')
# - Save to an output folder to avoid overwriting originals (you can switch to in-place)

import os
import re
import pandas as pd
import numpy as np

# --------- CONFIG: set exactly one of these ---------
# Option A: process all CSVs in a folder
folder_path = './'  # change to your folder with CSVs
process_all_csvs = True

# Option B: process a single CSV (set process_all_csvs = False)
file_path = './your_file.csv'

# Output directory (recommended)
output_dir = './cleaned_iso_dates'
os.makedirs(output_dir, exist_ok=True)

# --------- Heuristics to detect date-like columns safely ---------
def is_mostly_numeric_series(s: pd.Series) -> bool:
    s_nonnull = s.dropna()
    if s_nonnull.empty:
        return False
    # If dtype is numeric already, it's numeric
    if pd.api.types.is_numeric_dtype(s_nonnull):
        return True
    # For object, check if most values are numeric-looking
    cnt = len(s_nonnull)
    num_like = 0
    for v in s_nonnull.iloc[: min(500, cnt)]:  # sample up to 500
        if isinstance(v, (int, float, np.integer, np.floating)):
            num_like += 1
            continue
        if isinstance(v, str):
            vs = v.strip()
            # pure number or number with decimal
            if re.fullmatch(r'[+-]?\d+(\.\d+)?', vs):
                num_like += 1
    return num_like / cnt >= 0.8

def looks_like_date_string(v: str) -> bool:
    if not isinstance(v, str):
        return False
    s = v.strip()
    if s == '':
        return False
    # quick patterns for common date formats; do not include plain numbers (to avoid numeric IDs)
    patterns = [
        r'\d{4}[-/]\d{1,2}[-/]\d{1,2}',     # 2023-01-02 or 2023/1/2
        r'\d{1,2}[-/]\d{1,2}[-/]\d{4}',     # 1-2-2023 or 01/02/2023
        r'\d{1,2}\s+[A-Za-z]{3,}\s+\d{4}',  # 2 Jan 2023
        r'[A-Za-z]{3,}\s+\d{1,2},\s*\d{4}', # January 2, 2023
        r'\d{4}\s*[A-Za-z]{3,}\s*\d{1,2}',  # 2023 Jan 2
        r'\d{4}-\d{2}-\d{2}T',              # ISO datetime prefix
    ]
    for p in patterns:
        if re.search(p, s):
            return True
    return False

def is_date_like_series(s: pd.Series) -> bool:
    # Already datetime
    if pd.api.types.is_datetime64_any_dtype(s):
        return True
    # Exclude numeric or mostly numeric
    if is_mostly_numeric_series(s):
        return False
    s_nonnull = s.dropna()
    if s_nonnull.empty:
        return False
    # Sample to keep it fast
    sample = s_nonnull.astype(str).head(500)
    matches = sample.apply(looks_like_date_string).sum()
    ratio = matches / len(sample)
    return ratio >= 0.6  # majority looks like dates

def normalize_dates_in_df(df: pd.DataFrame) -> pd.DataFrame:
    # Identify candidate columns
    candidates = []
    for col in df.columns:
        try:
            if is_date_like_series(df[col]):
                candidates.append(col)
        except Exception:
            continue

    # Convert candidates safely
    for col in candidates:
        ser = df[col]
        # If already datetime, just format; else try to parse
        if not pd.api.types.is_datetime64_any_dtype(ser):
            ser_parsed = pd.to_datetime(ser, errors='coerce', infer_datetime_format=True, utc=False)
        else:
            ser_parsed = ser

        # If conversion produced very few valid dates, skip to avoid false positives
        nonnull = ser.notna().sum()
        valid = ser_parsed.notna().sum()
        if nonnull > 0 and valid / nonnull < 0.6:
            continue

        # Format as ISO date (YYYY-MM-DD); keep NaT as blank
        df[col] = ser_parsed.dt.strftime('%Y-%m-%d')
    return df

def process_one_csv(path_in: str, out_dir: str):
    try:
        df = pd.read_csv(path_in, low_memory=False)
    except Exception as e:
        print('Read failed: ' + path_in + ' | ' + str(e))
        return
    df2 = normalize_dates_in_df(df.copy())
    out_path = os.path.join(out_dir, os.path.basename(path_in))
    try:
        df2.to_csv(out_path, index=False)
        print('Saved: ' + out_path)
    except Exception as e:
        print('Save failed: ' + out_path + ' | ' + str(e))

# --------- Run ---------
if process_all_csvs:
    for fname in os.listdir(folder_path):
        if fname.lower().endswith('.csv'):
            process_one_csv(os.path.join(folder_path, fname), output_dir)
else:
    process_one_csv(file_path, output_dir)
```

```python
#Canonicalize city and campaign names.

# Canonicalize city and campaign names across all CSV files in a folder.
# - Uses your CITY_ALIASES and CAMPAIGN_ALIASES dictionaries
# - Detects likely city/campaign columns by name
# - Cleans text (strip, collapse spaces, remove accents)
# - City: title case unless alias maps to a specific form
# - Campaign: title case unless alias maps to a specific form
# - Overwrites the original CSVs in place

import os
import re
import unicodedata
import pandas as pd

folder_path = r'C:\Users\katka\OneDrive\ドキュメント\EDA_Project'

CITY_ALIASES = {
    'nyc': 'New York',
    'new york city': 'New York',
    'new york, ny': 'New York',
    'n y c': 'New York',
    'bombay': 'Mumbai',
    'bengaluru': 'Bengaluru',
    'bangalore': 'Bengaluru',
    'delhi': 'Delhi',
    'ncr': 'Delhi',
    'dilli': 'Delhi',
    'sfo': 'San Francisco',
    'san fran': 'San Francisco',
    'sf': 'San Francisco',
    'la': 'Los Angeles',
    'los angeles, ca': 'Los Angeles',
    'saint petersburg': 'St. Petersburg',
    'st petersburg': 'St. Petersburg',
    'st. petersburg': 'St. Petersburg',
    'saint louis': 'St. Louis',
    'st louis': 'St. Louis',
    'st. louis': 'St. Louis',
    'ho chi minh city': 'Ho Chi Minh City',
    'hcmc': 'Ho Chi Minh City',
    'hồ chí minh': 'Ho Chi Minh City',
    'sao paulo': 'São Paulo',
    'são paulo': 'São Paulo',
    'mexico df': 'Mexico City',
    'cdmx': 'Mexico City',
    'méxico d.f.': 'Mexico City',
}

CAMPAIGN_ALIASES = {
    'summer sale': 'Summer Sale',
    'summer_sale': 'Summer Sale',
    'summersale': 'Summer Sale',
    'black friday': 'Black Friday',
    'black_friday': 'Black Friday',
    'bf': 'Black Friday',
    'cyber monday': 'Cyber Monday',
    'cyber_monday': 'Cyber Monday',
    'cm': 'Cyber Monday',
    'new user acquisition': 'New User Acquisition',
    'new_user_acquisition': 'New User Acquisition',
    'nua': 'New User Acquisition',
    'retargeting': 'Retargeting',
    're-targeting': 'Retargeting',
    'rt': 'Retargeting',
    'brand awareness': 'Brand Awareness',
    'brand_awareness': 'Brand Awareness',
    'ba': 'Brand Awareness'
}


def strip_accents(text: str) -> str:
    """
    Remove diacritics from text for matching, preserving original for final output if needed.
    """
    if not isinstance(text, str):
        return text
    norm = unicodedata.normalize('NFKD', text)
    return ''.join([c for c in norm if not unicodedata.combining(c)])

def clean_key(text: str) -> str:
    """
    Prepare a string for alias lookup:
    - lowercase
    - remove diacritics
    - collapse whitespace
    - remove punctuation except spaces
    """
    if not isinstance(text, str):
        return text
    s = strip_accents(text).lower()
    s = re.sub(r'[^0-9a-z\s]+', ' ', s)
    s = re.sub(r'\s+', ' ', s).strip()
    return s

def to_title_keep_small_words(s: str) -> str:
    """
    Title-case while keeping common small words in lowercase (for readability).
    """
    if not isinstance(s, str):
        return s
    small = {'and', 'or', 'the', 'of', 'in', 'on', 'for', 'to', 'a', 'an'}
    tokens = re.sub(r'\s+', ' ', s.strip()).split(' ')
    titled = []
    for i, t in enumerate(tokens):
        if i == 0 or t.lower() not in small:
            titled.append(t.capitalize())
        else:
            titled.append(t.lower())
    return ' '.join(titled)

def canonicalize_value(val, alias_map, default_title=True):
    """
    Generic canonicalization using alias_map.
    - If a cleaned key exists in alias_map, return its mapped canonical form.
    - Else return title-cased (optional) cleaned original.
    """
    if pd.isna(val):
        return val
    original = str(val).strip()
    if original == '':
        return original
    key = clean_key(original)
    if key in alias_map:
        return alias_map[key]
    # If not in aliases, return title-cased original (normalized spaces)
    normalized = re.sub(r'\s+', ' ', original)
    return to_title_keep_small_words(normalized) if default_title else normalized

def is_city_col(colname: str) -> bool:
    """
    Heuristic to identify city-like columns.
    """
    return bool(re.search(r'(city|city_name|town|location_city|metro|municipality)', colname, flags=re.IGNORECASE))

def is_campaign_col(colname: str) -> bool:
    """
    Heuristic to identify campaign-like columns.
    """
    return bool(re.search(r'(campaign|campaign_name|mkt_campaign|utm_campaign|ad_campaign|promotion)', colname, flags=re.IGNORECASE))


for filename in os.listdir(folder_path):
    if not filename.lower().endswith('.csv'):
        continue

    file_path = os.path.join(folder_path, filename)
    try:
        df = pd.read_csv(file_path)
    except Exception as e:
        print('Skip (read error): ' + filename + ' | ' + str(e))
        continue

    # Detect columns
    city_cols = [c for c in df.columns if is_city_col(c)]
    campaign_cols = [c for c in df.columns if is_campaign_col(c)]

    # Apply canonicalization
    for c in city_cols:
        df[c] = df[c].apply(lambda x: canonicalize_value(x, CITY_ALIASES, default_title=True))
    for c in campaign_cols:
        df[c] = df[c].apply(lambda x: canonicalize_value(x, CAMPAIGN_ALIASES, default_title=True))

    # Save in place (change to an output folder if you prefer)
    try:
        df.to_csv(file_path, index=False)
        print('Updated: ' + filename + ' | Cities: ' + ', '.join(city_cols) + ' | Campaigns: ' + ', '.join(campaign_cols))
    except Exception as e:
        print('Save failed: ' + filename + ' | ' + str(e))
```

```python
# Standardize categorical values for status, payment_mode, and brand across all CSVs in a folder.
# - Detect likely columns by name
# - Normalize text (trim, collapse spaces, remove accents)
# - Apply alias maps to canonical values
# - Overwrite original files in place

import os
import re
import unicodedata
import pandas as pd

# --------- CONFIG ----------
folder_path = r'C:\Users\katka\OneDrive\ドキュメント\EDA_Project'  # change if needed

# Canonical mappings for status (expand per your data)
STATUS_ALIASES = {
    'pending': 'Pending',
    'in progress': 'Processing',
    'processing': 'Processing',
    'packed': 'Processing',
    'shipped': 'Shipped',
    'in transit': 'Shipped',
    'out for delivery': 'Out for Delivery',
    'delivered': 'Delivered',
    'completed': 'Delivered',
    'cancelled': 'Cancelled',
    'canceled': 'Cancelled',
    'returned': 'Returned',
    'refunded': 'Refunded',
    'failed': 'Failed',
    'payment failed': 'Failed',
    'on hold': 'On Hold'
}

# Canonical mappings for payment modes (expand per your data)
PAYMENT_MODE_ALIASES = {
    'cod': 'COD',
    'cash on delivery': 'COD',
    'cash-on-delivery': 'COD',
    'upi': 'UPI',
    'gpay': 'UPI',
    'google pay': 'UPI',
    'phonepe': 'UPI',
    'phone pe': 'UPI',
    'paytm upi': 'UPI',
    'net banking': 'Net Banking',
    'netbanking': 'Net Banking',
    'nb': 'Net Banking',
    'credit card': 'Credit Card',
    'creditcard': 'Credit Card',
    'cc': 'Credit Card',
    'debit card': 'Debit Card',
    'debitcard': 'Debit Card',
    'dc': 'Debit Card',
    'wallet': 'Wallet',
    'paytm wallet': 'Wallet',
    'mobikwik': 'Wallet',
    'freecharge': 'Wallet',
    'paypal': 'PayPal',
    'stripe': 'Stripe',
    'bank transfer': 'Bank Transfer',
    'banktransfer': 'Bank Transfer',
    'bt': 'Bank Transfer'
}

# Canonical mappings for brand names (expand per your catalog)
BRAND_ALIASES = {
    'apple inc.': 'Apple',
    'apple': 'Apple',
    'samsung electronics': 'Samsung',
    'samsung': 'Samsung',
    'oneplus': 'OnePlus',
    'one plus': 'OnePlus',
    'mi': 'Xiaomi',
    'xiaomi': 'Xiaomi',
    'redmi': 'Xiaomi',
    'google': 'Google',
    'google pixel': 'Google',
    'nothing': 'Nothing',
    'oppo': 'OPPO',
    'vivo': 'vivo',
    'realme': 'realme',
    'motorola': 'Motorola',
    'moto': 'Motorola',
    'huawei': 'Huawei',
    'honor': 'HONOR',
    'sony': 'Sony',
    'nokia': 'Nokia',
    'iqoo': 'iQOO',
    'asus': 'ASUS',
    'lenovo': 'Lenovo',
    'philips': 'Philips'
}

# --------- HELPERS ----------
def strip_accents_and_whitespace(text):
    if pd.isna(text):
        return text
    s = str(text).strip()
    s = re.sub(r'\s+', ' ', s)
    # normalize unicode and remove diacritics
    s_norm = unicodedata.normalize('NFKD', s)
    s_ascii = ''.join(ch for ch in s_norm if not unicodedata.combining(ch))
    return s_ascii

def keyify(text):
    if pd.isna(text):
        return None
    s = strip_accents_and_whitespace(text)
    s = s.lower()
    s = re.sub(r'[^0-9a-z]+', ' ', s).strip()
    s = re.sub(r'\s+', ' ', s)
    return s

def smart_title(text):
    if pd.isna(text):
        return text
    s = str(text)
    # keep common acronyms uppercase
    acronyms = set(['COD', 'UPI', 'EMI', 'GST', 'MRP', 'MSRP', 'LED', 'USB', 'NFC', '5G'])
    tokens = s.split(' ')
    fixed = []
    for t in tokens:
        t_clean = re.sub(r'[^0-9A-Za-z]+', '', t)
        if t_clean.upper() in acronyms:
            fixed.append(t_clean.upper())
        elif t_clean.lower() in ['and', 'or', 'of', 'the', 'in', 'for', 'on', 'by', 'to']:
            fixed.append(t_clean.lower())
        else:
            fixed.append(t_clean.capitalize())
    return ' '.join(fixed).strip()

def is_status_col(col):
    c = col.lower()
    return any(k in c for k in ['status', 'fulfillment', 'fulfilment', 'state'])

def is_payment_col(col):
    c = col.lower()
    return any(k in c for k in ['payment', 'pay_method', 'paymode', 'paymentmode', 'payment_type', 'pay type'])

def is_brand_col(col):
    c = col.lower()
    return any(k in c for k in ['brand', 'manufacturer', 'make', 'oem'])

def canonicalize_value(x, alias_map, default_title=True):
    if pd.isna(x):
        return x
    k = keyify(x)
    if k is None or k == '':
        return pd.NA
    # exact alias match
    if k in alias_map:
        return alias_map[k]
    # fallbacks for common status words joining
    k_compact = k.replace(' ', '')
    if k_compact in alias_map:
        return alias_map[k_compact]
    # otherwise prettify
    return smart_title(strip_accents_and_whitespace(x)) if default_title else strip_accents_and_whitespace(x)

# --------- PROCESS ALL CSVS ----------
for filename in os.listdir(folder_path):
    if not filename.lower().endswith('.csv'):
        continue
    file_path = os.path.join(folder_path, filename)
    try:
        df = pd.read_csv(file_path)
    except Exception as e:
        print('Skip (read error): ' + filename + ' | ' + str(e))
        continue

    # find candidate columns
    status_cols = [c for c in df.columns if is_status_col(c)]
    payment_cols = [c for c in df.columns if is_payment_col(c)]
    brand_cols = [c for c in df.columns if is_brand_col(c)]

    # apply canonicalization
    for c in status_cols:
        df[c] = df[c].apply(lambda v: canonicalize_value(v, STATUS_ALIASES, default_title=True))
    for c in payment_cols:
        df[c] = df[c].apply(lambda v: canonicalize_value(v, PAYMENT_MODE_ALIASES, default_title=True))
    for c in brand_cols:
        df[c] = df[c].apply(lambda v: canonicalize_value(v, BRAND_ALIASES, default_title=True))

    # save
    try:
        df.to_csv(file_path, index=False)
        print('Updated: ' + filename + ' | Status: ' + ', '.join(status_cols) + ' | Payment: ' + ', '.join(payment_cols) + ' | Brand: ' + ', '.join(brand_cols))
    except Exception as e:
        print('Save failed: ' + filename + ' | ' + str(e))
```

```python
import pandas as pd
import numpy as np

files = ["Customers.csv", "Customers_with_names_split.csv"]

invalid_values = [
    "", " ", "-", "--", "---", ".", 
    "nan", "NaN", "NULL", "null", 
    "None", "none", "N/A", "n/a"
]

for file in files:
    df = pd.read_csv(file)

    print("Processing:", file)

    # -------- EMAIL CLEANING --------
    if 'email' in df.columns:
        df['email'] = df['email'].astype(str).str.strip()
        df['email'] = df['email'].replace(invalid_values, np.nan)
        df['email'] = df['email'].fillna("NA")
    else:
        print("Warning: 'email' column not found in", file)

    # -------- PHONE CLEANING --------
    if 'phone' in df.columns:
        df['phone'] = df['phone'].astype(str).str.strip()
        df['phone'] = df['phone'].str.replace('.0', '', regex=False)
        df['phone'] = df['phone'].str.extract(r'(\d+)')

        # Convert invalid to NA
        df.loc[df['phone'].isin([None, "", " ", "nan", "NaN"]), 'phone'] = "NA"

        # Keep only 10-digit numbers
        df.loc[df['phone'].str.len() != 10, 'phone'] = "NA"
    else:
        print("Warning: 'phone' column not found in", file)

    # SAVE UPDATED FILE
    df.to_csv(file, index=False)
    print("Updated:", file, "\n")
```

```python
import pandas as pd
import numpy as np

sales = pd.read_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\Sales.csv")

# MODEL-WISE MEDIAN
model_median = sales.groupby('model_id')['discount'].median()

# IMPUTATION USING ONLY MODEL-WISE MEDIAN
def fill_discount(row):
    if pd.isna(row['discount']):
        return model_median[row['model_id']]   # only model median
    return row['discount']

sales['discount'] = sales.apply(fill_discount, axis=1)

print("Missing discount after model-wise fill:", sales['discount'].isnull().sum())

sales.to_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\Sales.csv", index=False)
```

```python
import pandas as pd
import numpy as np

# Load datasets
sales = pd.read_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\Sales.csv")
models = pd.read_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\MobileModels.csv")

# ------------------------------------------------------------
# We fill missing final_price using the REAL retail formula:
# final_price = base_price * quantity * (1 - discount/100)
#
# Reason:
# This method exactly matches how billing systems calculate the 
# price the customer actually pays after discount. 
# It is the most accurate and business-correct way to fix missing values.
# ------------------------------------------------------------

# Merge to bring in base model price
sales = sales.merge(models[['model_id', 'price']], on='model_id', how='left')

# Recalculate only where final_price is missing
def fix_final_price(row):
    if pd.isna(row['final_price']):
        return row['price'] * row['quantity'] * (1 - row['discount'] / 100)
    return row['final_price']

sales['final_price'] = sales.apply(fix_final_price, axis=1)

# Round to nearest rupee
sales['final_price'] = sales['final_price'].round(0)

# Drop temporary base price column
sales = sales.drop(columns=['price'])

# Save cleaned file
sales.to_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\Sales.csv", index=False)

print("Missing final_price values successfully recalculated and saved!")
```

```python
#Filling missing stock vlues with median
import pandas as pd
import numpy as np

stocks = pd.read_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\Stocks.csv")

# Clean column names
stocks.columns = stocks.columns.str.strip().str.lower()

# 1) Convert blanks (if any) to NaN
stocks['available_units'] = stocks['available_units'].replace(["", " "], np.nan)

# 2) Model+Shop wise median
group_median = stocks.groupby(['shop_id', 'model_id'])['available_units'].median()

# 3) Overall median fallback
overall_median = stocks['available_units'].median()

# 4) Impute missing values
def fill_stock(row):
    if pd.isna(row['available_units']):
        m = group_median.get((row['shop_id'], row['model_id']), np.nan)
        return overall_median if pd.isna(m) else m
    return row['available_units']

stocks['available_units'] = stocks.apply(fill_stock, axis=1)

# 5) Round final units
stocks['available_units'] = stocks['available_units'].round(0)

# Save back
stocks.to_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\Stocks.csv", index=False)

print("Missing stock units filled successfully!")
```

```python
#Checking and correcting outliers for all brands
import pandas as pd

# Load models
models = pd.read_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\MobileModels.csv")

# Step 1: Identify unrealistic low prices (< 6000)
wrong_prices = models[models["price"] < 6000]
print("Suspicious models:\n", wrong_prices)

# Pre-calc medians
ram_storage_median = models.groupby(["ram_gb", "storage_gb"])["price"].median()
brand_median = models.groupby("brand")["price"].median()
overall_median = models["price"].median()

# Step 2: Fix prices using hybrid method
for idx, row in wrong_prices.iterrows():
    
    ram = row["ram_gb"]
    storage = row["storage_gb"]
    brand = row["brand"]
    
    # 1) Try RAM + Storage median
    try:
        correct_price = ram_storage_median.loc[(ram, storage)]
    except:
        correct_price = None
    
    # If RAM+Storage not valid → use brand median
    if pd.isna(correct_price) or correct_price <= 0:
        correct_price = brand_median.loc[brand]
    
    # If brand also fails → fallback to overall median
    if pd.isna(correct_price) or correct_price <= 0:
        correct_price = overall_median
    
    # Round to integer
    correct_price = int(round(correct_price))
    
    # Assign fixed price
    models.loc[idx, "price"] = correct_price

# Save corrected models file
models.to_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\MobileModels.csv", index=False)

print("\nAll suspicious model prices have been corrected successfully!")
```

```python
import pandas as pd
import numpy as np

models_fp = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\MobileModels.csv"
sales_fp  = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\Sales.csv"

models = pd.read_csv(models_fp)
sales = pd.read_csv(sales_fp)

# ------------- 1) vivo models overview ----------
vivo_models = models[models['brand'].str.lower() == 'vivo']
print("VIVO MODELS COUNT:", len(vivo_models))
print(vivo_models[['model_id','ram_gb','storage_gb','price']].sort_values('price').to_string(index=False))

print("\nVIVO price summary:")
print(vivo_models['price'].describe())

# ------------- 2) Compare with same RAM+Storage across brands ----------
# For the suspect model MOD2199 find its specs
suspect = models[models['model_id']=='MOD2199']
if suspect.empty:
    print("\nMOD2199 not found in models file.")
else:
    ram = int(suspect['ram_gb'].iloc[0])
    storage = int(suspect['storage_gb'].iloc[0])
    print(f"\nMOD2199 specs: ram_gb={ram}, storage_gb={storage}, current_price={suspect['price'].iloc[0]}")

    same_specs = models[(models['ram_gb']==ram) & (models['storage_gb']==storage)]
    print(f"\nNumber of models with same specs (ram {ram}, storage {storage}):", len(same_specs))
    print(same_specs[['model_id','brand','price']].sort_values('price').to_string(index=False))

    print("\nSame-specs price summary:")
    print(same_specs['price'].describe())

# ------------- 3) Find all suspicious low-price models (threshold adjustable) ----------
threshold = 6000   # change if you want a different cutoff
suspects_all = models[models['price'] < threshold].sort_values('price')
print(f"\nModels with price < {threshold} (suspicious):")
print(suspects_all[['model_id','brand','ram_gb','storage_gb','price']].to_string(index=False))

# ------------- 4) Suggest replacement price per model using RAM+Storage median ----------
# compute median by ram+storage (exclude non-positive prices)
group_median = models[models['price']>0].groupby(['ram_gb','storage_gb'])['price'].median()

def suggest_price(row):
    key = (row['ram_gb'], row['storage_gb'])
    if key in group_median.index:
        return int(round(group_median.loc[key]))
    else:
        # fallback to brand median
        bmedian = models[(models['brand']==row['brand']) & (models['price']>0)]['price'].median()
        return int(round(bmedian)) if pd.notna(bmedian) else int(round(models['price'].median()))

# show suggestions for suspicious models
if not suspects_all.empty:
    print("\nSuggested replacement prices (RAM+Storage median fallback brand):")
    suggestions = []
    for _, r in suspects_all.iterrows():
        suggested = suggest_price(r)
        suggestions.append((r['model_id'], r['brand'], r['ram_gb'], r['storage_gb'], r['price'], suggested))
    print(pd.DataFrame(suggestions, columns=['model_id','brand','ram_gb','storage_gb','current_price','suggested_price']).to_string(index=False))
else:
    print("\nNo suspicious low-price models found under threshold.")

# ------------- 5) OPTIONAL: Apply fixes automatically (UNCOMMENT to run) ----------
apply_fixes = False   # set True to apply fixes automatically
if apply_fixes and not suspects_all.empty:
    for idx, row in suspects_all.iterrows():
        new_price = suggest_price(row)
        models.loc[idx, 'price'] = new_price
    models.to_csv(models_fp, index=False)
    print("\nApplied fixes and saved MobileModels.csv")

    # Recalculate final_price for all sales and save
    models_ref = models[['model_id','price']]
    merged = sales.merge(models_ref, on='model_id', how='left')
    merged['final_price'] = (merged['price'] * merged['quantity'] * (1 - merged['discount']/100)).round(2)
    merged.to_csv(sales_fp, index=False)
    print("Recalculated and saved Sales.csv with new final_price.")
else:
    print("\nNot applying fixes automatically. To apply fixes set apply_fixes = True and re-run.")
```

```python
import pandas as pd
from ipywidgets import Dropdown, VBox, Output
from IPython.display import display

# Load dataset
models_fp = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\MobileModels.csv"
df = pd.read_csv(models_fp)

# Widgets
brand_dd = Dropdown(description="Brand:", options=["Select"] + sorted(df["brand"].unique()))
model_dd = Dropdown(description="Model:", options=["Select"])
ram_dd = Dropdown(description="RAM:", options=["Select"])
storage_dd = Dropdown(description="Storage:", options=["Select"])
output = Output()

# ---------------------- Update Functions ----------------------

def update_model_options(change):
    brand = change['new']
    if brand == "Select":
        model_dd.options = ["Select"]
        ram_dd.options = ["Select"]
        storage_dd.options = ["Select"]
        return
    
    models = df[df["brand"] == brand]["model_name"].unique()
    model_dd.options = ["Select"] + sorted(models)
    ram_dd.options = ["Select"]
    storage_dd.options = ["Select"]

def update_ram_options(change):
    brand = brand_dd.value
    model = change['new']
    if model == "Select":
        ram_dd.options = ["Select"]
        storage_dd.options = ["Select"]
        return
    
    rams = df[(df["brand"] == brand) & (df["model_name"] == model)]["ram_gb"].unique()
    ram_dd.options = ["Select"] + sorted(rams)
    storage_dd.options = ["Select"]

def update_storage_options(change):
    brand = brand_dd.value
    model = model_dd.value
    ram = change['new']
    if ram == "Select":
        storage_dd.options = ["Select"]
        return
    
    storages = df[
        (df["brand"] == brand) &
        (df["model_name"] == model) &
        (df["ram_gb"] == ram)
    ]["storage_gb"].unique()

    storage_dd.options = ["Select"] + sorted(storages)

def show_price(change):
    output.clear_output()
    brand = brand_dd.value
    model = model_dd.value
    ram = ram_dd.value
    storage = change['new']
    
    if storage == "Select":
        return
    
    result = df[
        (df["brand"] == brand) &
        (df["model_name"] == model) &
        (df["ram_gb"] == ram) &
        (df["storage_gb"] == storage)
    ]
    
    with output:
        if result.empty:
            print("❗ No mobile with this combination.")
        else:
            print("📌 MATCH FOUND:\n")
            display(result[["brand","model_name","ram_gb","storage_gb","price"]])

# ---------------------- Link Dropdowns ----------------------

brand_dd.observe(update_model_options, names="value")
model_dd.observe(update_ram_options, names="value")
ram_dd.observe(update_storage_options, names="value")
storage_dd.observe(show_price, names="value")

# ---------------------- Display UI ----------------------

display(VBox([brand_dd, model_dd, ram_dd, storage_dd, output]))
```

```python

#Replace all prices <10k with median of same specification range
import pandas as pd

models_fp = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\MobileModels.csv"

# Load data
models = pd.read_csv(models_fp)

# -------- STEP 1: Identify all models below 10k --------
below_10k = models[models["price"] < 10000].copy()

print("\n✔ Total models <10k:", len(below_10k))

# -------- STEP 2: Compute median price for each (ram, storage) combination --------
median_prices = (
    models.groupby(["ram_gb", "storage_gb"])["price"]
    .median()
    .reset_index()
    .rename(columns={"price": "median_price"})
)

print("\n✔ Median prices calculated for each specification group:\n")
print(median_prices.to_string(index=False))

# -------- STEP 3: Merge median values back into original dataframe --------
models = models.merge(
    median_prices,
    on=["ram_gb", "storage_gb"],
    how="left"
)

# -------- STEP 4: Replace price <10k with median of that spec group --------
models["price"] = models.apply(
    lambda row: row["median_price"] if row["price"] < 10000 else row["price"],
    axis=1
)

print("\n✔ Replacement complete! All values <10k have been updated.\n")

# -------- STEP 5: Optional – save cleaned version --------
output_fp = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\MobileModels_Cleaned.csv"
models.to_csv(output_fp, index=False)

print("✔ Cleaned file saved to:")
print(output_fp)
```

```python
import pandas as pd

models_fp = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\MobileModels.csv"

# Load data
models = pd.read_csv(models_fp)

# Models costing <10k
cheap = models[models["price"] < 10000].copy()

print("\n✔ Models below 10k:", len(cheap))


# -------- STEP 1: Function to compute median from other brands --------
def get_other_brand_median(ram, storage, brand):
    # same RAM + storage
    same_spec = models[(models["ram_gb"] == ram) & (models["storage_gb"] == storage)]
    
    # remove same brand
    other_brands = same_spec[same_spec["brand"] != brand]
    
    if len(other_brands) == 0:
        return None  # no other brands available
    
    return other_brands["price"].median()


# -------- STEP 2: Replace each cheap model’s price --------
for idx, row in cheap.iterrows():
    ram  = row["ram_gb"]
    stor = row["storage_gb"]
    brand = row["brand"]
    old_price = row["price"]

    median_other = get_other_brand_median(ram, stor, brand)

    if median_other is not None:
        models.loc[idx, "price"] = median_other
        print(f"\nUPDATED: {row['model_id']} | {brand} | {ram}/{stor}GB")
        print(f"Old price: {old_price}")
        print(f"New price (median from other brands): {median_other}")
    else:
        print(f"\nSKIPPED: {row['model_id']} (no other brands with same spec)")


# -------- STEP 3: Save updated file --------
output_fp = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\MobileModels_Fixed.csv"
models.to_csv(output_fp, index=False)

print("\n✔ All updates completed.")
print("✔ Cleaned file saved at:")
print(output_fp)
```

```python
import pandas as pd
from ipywidgets import Dropdown, VBox, Output
from IPython.display import display

# ---------------- LOAD FILES ----------------
models_fp = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\MobileModels.csv"
sales_fp  = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\Sales.csv"
stocks_fp = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\Stocks.csv"

models = pd.read_csv(models_fp)
sales  = pd.read_csv(sales_fp)
stocks = pd.read_csv(stocks_fp)

# Merge all datasets for complete info
merged = models.merge(stocks, on="model_id", how="left") \
               .merge(sales[["model_id", "discount", "base_price"]], 
                      on="model_id", how="left")

# ---------- USER INTERFACE WIDGETS ----------
brand_dropdown = Dropdown(
    description="Brand:",
    options=["Select a brand"] + sorted(merged["brand"].unique()),
    value="Select a brand"
)

model_dropdown = Dropdown(
    description="Model:",
    options=["Select a model"],
    value="Select a model"
)

output = Output()


# ---------- UPDATE MODEL OPTIONS WHEN BRAND SELECTED ----------
def update_models(change):
    if change['new'] == "Select a brand":
        model_dropdown.options = ["Select a model"]
        return
    
    brand = change['new']
    brand_models = merged[merged["brand"] == brand]["model_id"].unique()
    
    model_dropdown.options = ["Select a model"] + list(brand_models)


brand_dropdown.observe(update_models, names="value")


# ---------- DISPLAY MODEL DETAILS ----------
def show_model_details(change):
    output.clear_output()

    if change['new'] == "Select a model":
        return
    
    model_id = change['new']
    row = merged[merged["model_id"] == model_id].iloc[0]

    with output:
        print("\n📱 MODEL DETAILS")
        print("------------------------------------")
        print(f"Brand:           {row['brand']}")
        print(f"Model Name:      {row['model_name']}")
        print(f"Model ID:        {row['model_id']}")
        print(f"RAM:             {row['ram_gb']} GB")
        print(f"Storage:         {row['storage_gb']} GB")
        print(f"Base Price:      ₹{row['base_price']:.2f}")
        print(f"Discount:        {row['discount']} %")
        print(f"Available Stock: {row['available_units']}")
        print("------------------------------------")


model_dropdown.observe(show_model_details, names="value")


# ---------- DISPLAY THE FULL INTERFACE ----------
display(VBox([brand_dropdown, model_dropdown, output]))
```

```python
#round final price of device upto 2 decimals

import pandas as pd

# Sales file path
sales_fp = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\Sales.csv"

# Load Sales data
sales = pd.read_csv(sales_fp)

print("🔄 Recalculating final_price with rounding...")

# Ensure discount is numeric
sales["discount"] = pd.to_numeric(sales["discount"], errors="coerce").fillna(0)

# Recalculate final_price (per unit price * quantity)
sales["final_price"] = (
    sales["quantity"] * sales["base_price"] * (1 - sales["discount"] / 100)
).round(2)     # ✅ Limit to 2 decimals

print("✔ final_price recalculated and rounded to 2 decimals.")

# Save back into SAME CSV (overwrite old file)
sales.to_csv(sales_fp, index=False)

print("✔ All changes saved successfully to Sales.csv")
```

```python
import pandas as pd
import plotly.express as px
from ipywidgets import Dropdown, HBox, VBox, Output
from IPython.display import display

# Load your data
sales = pd.read_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\Sales.csv")
models = pd.read_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\MobileModels.csv")
```

```python
# Dynamic Price Distribution (Brand / RAM / Storage Filters)

# Filter dropdowns
brand_f = Dropdown(
    description="Brand", 
    options=["All"] + sorted(models["brand"].unique())
)

ram_f = Dropdown(
    description="RAM", 
    options=["All"] + sorted(models["ram_gb"].unique())
)

stor_f = Dropdown(
    description="Storage", 
    options=["All"] + sorted(models["storage_gb"].unique())
)

out2 = Output()

def update_price_graph(change):
    out2.clear_output()
    df = models.copy()

    # Apply filters
    if brand_f.value != "All":
        df = df[df["brand"] == brand_f.value]

    if ram_f.value != "All":
        df = df[df["ram_gb"] == ram_f.value]

    if stor_f.value != "All":
        df = df[df["storage_gb"] == stor_f.value]

    # If no data left after filtering → show message
    if df.empty:
        with out2:
            print("No data available for the selected filters.")
        return

    # Histogram
    fig_hist = px.histogram(
        df, 
        x="price", 
        nbins=30, 
        title="Price Distribution",
        labels={"price": "Price"}
    )

    # Boxplot
    fig_box = px.box(
        df, 
        y="price", 
        title="Price Boxplot",
        labels={"price": "Price"}
    )

    with out2:
        display(fig_hist)
        display(fig_box)

# Trigger graph update whenever filters change
brand_f.observe(update_price_graph, names='value')
ram_f.observe(update_price_graph, names='value')
stor_f.observe(update_price_graph, names='value')

# Display panel
display(VBox([HBox([brand_f, ram_f, stor_f]), out2]))
```

```python
#Dynamic Shop Performance Dashboard
shop_dd = Dropdown(description="Shop", options=sorted(sales["shop_id"].unique()))
out3 = Output()

def show_shop_dashboard(change):
    out3.clear_output()
    shop = change['new']
    
    shop_sales = sales[sales["shop_id"] == shop].copy()
    total_sales = shop_sales["quantity"].sum()
    revenue = (shop_sales["final_price"]).sum()
    best_model = shop_sales.groupby("model_id")["quantity"].sum().idxmax()

    stock_shop = stocks[stocks["shop_id"] == shop]
    
    fig = px.bar(stock_shop, x="model_id", y="available_units",
                 title=f"Stock Availability in {shop}")

    with out3:
        print(f"📌 Shop ID: {shop}")
        print(f"🛒 Total Units Sold: {total_sales}")
        print(f"💰 Total Revenue: ₹{revenue:,.2f}")
        print(f"⭐ Best Selling Model: {best_model}")
        display(fig)

shop_dd.observe(show_shop_dashboard, names='value')
display(VBox([shop_dd, out3]))
```

```python
# ----------------------------------------------------------
# STEP 1: IMPORT LIBRARIES AND LOAD THE DATA
# ----------------------------------------------------------

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
import matplotlib.pyplot as plt

# Load your datasets
sales  = pd.read_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\Sales.csv")
models = pd.read_csv(r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project\MobileModels.csv")

# Merge sales with model details
df = sales.merge(models, on="model_id", how="left")

df.head()
```

```python
# ----------------------------------------------------------
# STEP 2: CREATE TARGET VARIABLE (Low / Medium / High Sales)
# ----------------------------------------------------------

df["sales_category"] = pd.cut(
    df["quantity"],
    bins=[-1, 20, 50, 999999],     # Ranges for low, med, high
    labels=["Low", "Medium", "High"]
)

df[["quantity", "sales_category"]].head()
```

```python
# ----------------------------------------------------------
# STEP 1: BUILD CLEAN ML DATAFRAME
# ----------------------------------------------------------

# Use the correct model info columns (the _y ones)
df_ml = df[[
    "brand_y",
    "ram_gb_y",
    "storage_gb_y",
    "price",
    "sales_category"
]].copy()

# Rename them for easier use
df_ml.rename(columns={
    "brand_y": "brand",
    "ram_gb_y": "ram_gb",
    "storage_gb_y": "storage_gb"
}, inplace=True)

print(df_ml.head())
```

```python
# ----------------------------------------------------------
# STEP 3: SELECT FEATURES AND TARGET
# ----------------------------------------------------------

X = df_ml[["brand", "ram_gb", "storage_gb", "price"]]
y = df_ml["sales_category"]
```

```python
df_ml = df[["brand_y", "ram_gb_y", "storage_gb_y", "price", "sales_category"]]
```

```python
# STEP 4 — LABEL ENCODING (NO WARNING VERSION)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# Use .loc to avoid warnings
X.loc[:, "brand"] = le.fit_transform(X["brand"])
```

```python
# ----------------------------------------------------------
# STEP 5: TRAIN–TEST SPLIT
# ----------------------------------------------------------

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,                  # features
    y,                  # target
    test_size=0.2,      # 20% test data
    random_state=42     # ensure reproducibility
)
```

```python
# ----------------------------------------------------------
# STEP 6: TRAIN DECISION TREE CLASSIFIER
# ----------------------------------------------------------

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(
    max_depth=5,     # to avoid overfitting
    random_state=42
)

model.fit(X_train, y_train)

print("Model training completed successfully ✔️")
```

```python
# ----------------------------------------------------------
# STEP 7: EVALUATE THE MODEL
# ----------------------------------------------------------

from sklearn.metrics import accuracy_score, classification_report

# Predict on test data
y_pred = model.predict(X_test)

# Accuracy
print("🔍 Model Accuracy:", accuracy_score(y_test, y_pred))

# Detailed performance
print("\n📊 Classification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd

folder = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project"

# Load all datasets
customers = pd.read_csv(folder + "\\Customers.csv")
models    = pd.read_csv(folder + "\\MobileModels.csv")
shops     = pd.read_csv(folder + "\\ShopLocations.csv")
sales     = pd.read_csv(folder + "\\Sales.csv")
offers    = pd.read_csv(folder + "\\Offers.csv")
stocks    = pd.read_csv(folder + "\\Stocks.csv")

# -------------------------------------------------------------------
# 1) FIX ORPHAN KEYS (so merges don't produce nulls)
# -------------------------------------------------------------------
sales["customer_id"] = sales["customer_id"].where(
    sales["customer_id"].isin(customers["customer_id"]), "UNKNOWN_CUST"
)

sales["model_id"] = sales["model_id"].where(
    sales["model_id"].isin(models["model_id"]), "UNKNOWN_MODEL"
)

sales["shop_id"] = sales["shop_id"].where(
    sales["shop_id"].isin(shops["shop_id"]), "UNKNOWN_SHOP"
)

sales["offer_id"] = sales.get("offer_id", pd.Series(["NO_OFFER"] * len(sales)))

# -------------------------------------------------------------------
# 2) MERGE IN CORRECT ORDER
# -------------------------------------------------------------------

# Step A — Sales + Customers
unified = sales.merge(customers, on="customer_id", how="left")

# Step B — Add model details
unified = unified.merge(
    models,
    on="model_id",
    how="left",
    suffixes=("", "_model")
)

# Step C — Add shop details
unified = unified.merge(
    shops.rename(columns={"location": "shop_location"}),
    on="shop_id",
    how="left",
    suffixes=("", "_shop")
)

# Step D — Add offer details
unified = unified.merge(
    offers,
    on=["model_id", "shop_id"],
    how="left",
    suffixes=("", "_offer")
)

# Step E — Add stock availability
unified = unified.merge(
    stocks[["shop_id", "model_id", "available_units"]],
    on=["shop_id", "model_id"],
    how="left"
)

# -------------------------------------------------------------------
# 3) SAVE FINAL FILE
# -------------------------------------------------------------------
output_path = folder + "\\Unified_Schema.csv"
unified.to_csv(output_path, index=False)

print("\n✔ Unified schema created successfully!")
print(f"Saved to: {output_path}")
print(f"Final shape: {unified.shape}")

print("\nSample rows:")
print(unified.head())
```

```python
#2.Orphan key resolution code:
import pandas as pd

folder = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project"

# Load all datasets
customers = pd.read_csv(folder + "\\Customers.csv")
models = pd.read_csv(folder + "\\MobileModels.csv")
shops = pd.read_csv(folder + "\\ShopLocations.csv")
sales = pd.read_csv(folder + "\\Sales.csv")
offers = pd.read_csv(folder + "\\Offers.csv")
stocks = pd.read_csv(folder + "\\Stocks.csv")

print("Files loaded.")


# ======================================================
# 1️⃣ FIX ORPHAN KEYS IN SALES
# ======================================================

# Replace invalid customer_id with safe dummy ID
sales["customer_id"] = sales["customer_id"].where(
    sales["customer_id"].isin(customers["customer_id"]),
    "CUST_UNKNOWN"
)

# Replace invalid model_id with safe dummy ID
sales["model_id"] = sales["model_id"].where(
    sales["model_id"].isin(models["model_id"]),
    "MOD_UNKNOWN"
)

# Replace invalid shop_id with safe dummy ID
sales["shop_id"] = sales["shop_id"].where(
    sales["shop_id"].isin(shops["shop_id"]),
    "SHOP_UNKNOWN"
)

# Replace invalid offer_id with safe dummy ID
if "offer_id" in sales.columns:
    sales["offer_id"] = sales["offer_id"].where(
        sales["offer_id"].isin(offers["offer_id"]),
        "OFF_UNKNOWN"
    )
else:
    # Sales had no offer_id → create one
    sales["offer_id"] = "OFF_UNKNOWN"



# ======================================================
# 2️⃣ FIX ORPHAN KEYS IN OFFERS
# ======================================================

offers["model_id"] = offers["model_id"].where(
    offers["model_id"].isin(models["model_id"]),
    "MOD_UNKNOWN"
)

offers["shop_id"] = offers["shop_id"].where(
    offers["shop_id"].isin(shops["shop_id"]),
    "SHOP_UNKNOWN"
)



# ======================================================
# 3️⃣ FIX ORPHAN KEYS IN STOCKS
# ======================================================

stocks["model_id"] = stocks["model_id"].where(
    stocks["model_id"].isin(models["model_id"]),
    "MOD_UNKNOWN"
)

stocks["shop_id"] = stocks["shop_id"].where(
    stocks["shop_id"].isin(shops["shop_id"]),
    "SHOP_UNKNOWN"
)



# ======================================================
# SAVE FIXED FILES
# ======================================================
sales.to_csv(folder + "\\Sales.csv", index=False)
offers.to_csv(folder + "\\Offers.csv", index=False)
stocks.to_csv(folder + "\\Stocks.csv", index=False)

print("\n✔ ORPHAN KEYS FIXED FOR:")
print("   - Sales")
print("   - Offers")
print("   - Stocks")

print("\n✔ Dummy IDs used:")
print("   CUST_UNKNOWN")
print("   MOD_UNKNOWN")
print("   SHOP_UNKNOWN")
print("   OFF_UNKNOWN")
```

```python
import pandas as pd

folder = r"C:\Users\katka\OneDrive\ドキュメント\EDA_Project"

# Load cleaned files
customers = pd.read_csv(folder + "\\Customers.csv")
models = pd.read_csv(folder + "\\MobileModels.csv")
shops = pd.read_csv(folder + "\\ShopLocations.csv")
sales = pd.read_csv(folder + "\\Sales.csv")
offers = pd.read_csv(folder + "\\Offers.csv")
stocks = pd.read_csv(folder + "\\Stocks.csv")

print("All cleaned files loaded.")


# ======================================================
# FIX: MobileModels uses column name  "price"
# So rename it to base_price BEFORE merging
# ======================================================
models = models.rename(columns={"price": "base_price"})

model_cols = ["model_id", "brand", "model_name", "ram_gb", "storage_gb", "base_price"]


# ======================================================
# STEP 1 — MERGE SALES + CUSTOMERS
# ======================================================
unified = sales.merge(
    customers,
    on="customer_id",
    how="left"
)


# ======================================================
# STEP 2 — MERGE MODELS
# ======================================================
unified = unified.merge(
    models[model_cols],
    on="model_id",
    how="left"
)


# ======================================================
# STEP 3 — MERGE SHOPS
# ======================================================
shops = shops.rename(columns={"location": "shop_location"})

unified = unified.merge(
    shops[["shop_id", "shop_location"]],
    on="shop_id",
    how="left"
)


# ======================================================
# STEP 4 — MERGE OFFERS
# ======================================================
unified = unified.merge(
    offers[["offer_id", "discount_percent", "start_date", "end_date", "campaign_name"]],
    on="offer_id",
    how="left"
)


# ======================================================
# STEP 5 — MERGE STOCKS
# ======================================================
unified = unified.merge(
    stocks[["model_id", "shop_id", "available_units"]],
    on=["model_id", "shop_id"],
    how="left"
)


# ======================================================
# STEP 6 — SAVE UNIFIED SCHEMA
# ======================================================
save_path = folder + "\\Unified_Schema.csv"
unified.to_csv(save_path, index=False)

print("\n✔ Unified schema created and saved!")
print("Path:", save_path)

print("\nUnified shape:", unified.shape)
print("\nSample rows:\n")
print(unified.head())
```
