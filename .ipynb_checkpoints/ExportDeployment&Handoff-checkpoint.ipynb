```python
# -------------------------------------------------------------
# LOAD ALL CLEANED DATASETS BEFORE BUILDING STAR SCHEMA
# -------------------------------------------------------------
import pandas as pd

customers = pd.read_csv("Customers.csv")
models = pd.read_csv("MobileModels.csv")
shops = pd.read_csv("ShopLocations.csv")
sales = pd.read_csv("Sales.csv")
stocks = pd.read_csv("Stocks.csv")
offers = pd.read_csv("Offers.csv")

print("All datasets loaded successfully!")
```

```python
# -------------------------------------------------------------
# STEP 10.1: BUILD STAR SCHEMA (FACT + DIM TABLES)
# -------------------------------------------------------------
import pandas as pd
import os

# Ensure output folder
os.makedirs("exports", exist_ok=True)

# Dimension Tables
DimCustomer = customers.copy()
DimModel = models.copy()
DimShop = shops.copy()
DimOffer = offers.copy()
DimStock = stocks.copy()

# Fact Table (Sales)
FactSales = sales.copy()

# Add revenue column if needed
FactSales["revenue"] = FactSales["quantity"] * FactSales["final_price"]

print("Star Schema Tables Created:")
print("→ DimCustomer")
print("→ DimModel")
print("→ DimShop")
print("→ DimOffer")
print("→ DimStock")
print("→ FactSales (with revenue)")
```
