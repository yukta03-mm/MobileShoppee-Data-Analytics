```python
# Import Required Libraries
import pandas as pd
import numpy as np

# Load All CSV Files
customers = pd.read_csv("Customers.csv")
models = pd.read_csv("MobileModels.csv")
shops = pd.read_csv("ShopLocations.csv")
sales = pd.read_csv("Sales.csv")
stocks = pd.read_csv("Stocks.csv")
offers = pd.read_csv("Offers.csv")

# Preview
print("Data Loaded Successfully!\n")
print("Sales Columns:", sales.columns.tolist())
print("Customers Columns:", customers.columns.tolist())
```

```python
# derive KPIs

sales["revenue"] = sales["quantity"] * sales["final_price"]
total_revenue = sales["revenue"].sum()
print(f"Total Revenue: ₹{total_revenue:,.2f}")
```

```python
avg_discount = sales["discount"].mean()
print(f"🏷 Average Discount: {avg_discount:.2f}%")
```

```python
repeat_customers = sales["customer_id"].value_counts()
num_repeat_customers = (repeat_customers > 1).sum()
total_customers = sales["customer_id"].nunique()
repeat_ratio = (num_repeat_customers / total_customers) * 100
print(f" Repeat Customers: {num_repeat_customers} ({repeat_ratio:.2f}% of all customers)")
```

```python
# Merge sales and stock data
stock_merge = pd.merge(stocks, sales.groupby("model_id")["quantity"].sum().reset_index(), 
                       on="model_id", how="left")

stock_merge["stock_turnover"] = stock_merge["quantity"] / stock_merge["available_units"]
avg_turnover = stock_merge["stock_turnover"].mean()
print(f"Average Stock Turnover: {avg_turnover:.2f}")
```

```python
# Merge offers with sales on common columns: model_id and shop_id
sales_offers = pd.merge(sales, offers, on=["model_id", "shop_id"], how="left")

# Ensure 'revenue' exists
sales_offers["revenue"] = sales_offers["quantity"] * sales_offers["final_price"]

# Group by campaign or offer_id (if present)
roi_df = sales_offers.groupby("campaign_name").agg({
    "revenue": "sum",
    "discount_percent": "mean"
}).reset_index()

# Compute ROI estimate safely
roi_df["roi_estimate"] = roi_df["revenue"] / (roi_df["discount_percent"] + 1e-6)

# Display
print("Campaign ROI (Top 5 offers):")
print(roi_df.head())
```
