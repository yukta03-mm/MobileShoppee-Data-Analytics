```python
#10.1 Export curated Fact & Dim tables to CSV/Excel.
```

```python
import pandas as pd
import os

# Create output folder
output_folder = "exported_tables"
os.makedirs(output_folder, exist_ok=True)

# Load curated Fact & Dimension tables
fact_sales = pd.read_csv("FactSales.csv")
dim_customer = pd.read_csv("DimCustomer.csv")
dim_model = pd.read_csv("DimModel.csv")
dim_offer = pd.read_csv("DimOffer.csv")
dim_shop = pd.read_csv("DimShop.csv")
dim_stock = pd.read_csv("DimStock.csv")

print("Loaded all curated tables successfully!")
```

```python
# Export curated tables to CSV
fact_sales.to_csv(f"{output_folder}/FactSales.csv", index=False)
dim_customer.to_csv(f"{output_folder}/DimCustomer.csv", index=False)
dim_model.to_csv(f"{output_folder}/DimModel.csv", index=False)
dim_offer.to_csv(f"{output_folder}/DimOffer.csv", index=False)
dim_shop.to_csv(f"{output_folder}/DimShop.csv", index=False)
dim_stock.to_csv(f"{output_folder}/DimStock.csv", index=False)

print("CSV export completed successfully!")
```

```python
excel_path = f"{output_folder}/Curated_Tables.xlsx"

# Export all curated tables to a single Excel file
with pd.ExcelWriter(excel_path, engine='xlsxwriter') as writer:
    fact_sales.to_excel(writer, sheet_name="FactSales", index=False)
    dim_customer.to_excel(writer, sheet_name="DimCustomer", index=False)
    dim_model.to_excel(writer, sheet_name="DimModel", index=False)
    dim_offer.to_excel(writer, sheet_name="DimOffer", index=False)
    dim_shop.to_excel(writer, sheet_name="DimShop", index=False)
    dim_stock.to_excel(writer, sheet_name="DimStock", index=False)

print("Excel export completed successfully!")
```
